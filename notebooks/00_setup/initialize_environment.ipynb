```python
# Define database
db_name = "ecommerce_lakehouse"

# Create database
spark.sql(f"CREATE DATABASE IF NOT EXISTS {db_name}")

# Use database
spark.sql(f"USE {db_name}")

print("Lakehouse database initialized:", db_name)
```

```python
# Create Volume Properly 
spark.sql("USE CATALOG workspace")

spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.ecommerce_lakehouse")
# Creating Volume
spark.sql("""
CREATE VOLUME IF NOT EXISTS workspace.ecommerce_lakehouse.raw_events_volume
""")
# verify volume
display(dbutils.fs.ls("/Volumes/workspace/ecommerce_lakehouse/raw_events_volume"))
```
