## ** Generate Synthetic Clickstream Events**

```python
#generator code:
import json
import random
import time
import os

output_path = "/Volumes/workspace/ecommerce_lakehouse/raw_events_volume/event_batches"

os.makedirs(output_path, exist_ok=True)

users = [f"U100{i}" for i in range(1,11)]
products = [f"P10{i}" for i in range(1,11)]

event_types = ["view","add_to_cart","purchase"]

countries = ["India","USA","Germany","UK","Canada"]
devices = ["mobile","desktop","tablet"]

for batch in range(1,21):

    events = []

    for i in range(200):

        event_type = random.choice(event_types)

        fraud_flag = 0
        if event_type == "purchase" and random.random() < 0.05:
            fraud_flag = 1

        event = {
            "user_id": random.choice(users),
            "session_id": f"S{random.randint(1000,9999)}",
            "product_id": random.choice(products),
            "event_type": event_type,
            "price": round(random.uniform(20,700),2),
            "device": random.choice(devices),
            "country": random.choice(countries),
            "fraud_flag": fraud_flag,
            "ts": int(time.time())
        }

        events.append(event)

    file_path = f"{output_path}/event_batch_{batch:03d}.json"

    with open(file_path,"w") as f:
        for e in events:
            f.write(json.dumps(e) + "\n")

print("Advanced synthetic event batches generated.")
```
