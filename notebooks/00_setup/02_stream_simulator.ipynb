```python
import time
import glob
import shutil
import os

SRC_DIR = "/Volumes/workspace/ecommerce_lakehouse/raw_events_volume/event_batches/"
DST_DIR = "/Volumes/workspace/ecommerce_lakehouse/raw_events_volume/stream_input/"

os.makedirs(DST_DIR, exist_ok=True)

files = sorted(glob.glob(SRC_DIR + "*.json"))

print("Starting streaming simulation...\n")

for f in files:

    target = DST_DIR + os.path.basename(f)

    shutil.copy(f, target)

    print("Streamed:", os.path.basename(f))

    time.sleep(5)
```
